```python
from __future__ import division
```

```python
train_tags = {}
all_tags = set()
with open("./rawInput/train.csv", "r") as train_labels:
    train_labels.readline()
    for line in train_labels:
        img, tags = line.rstrip("\n").split(",")
        train_tags[img] = tags.split(" ")
        all_tags.update(train_tags[img])
print train_tags[train_tags.keys()[0]]
```

```python
import math
count_tags = {}
for img in train_tags:
    tags = train_tags[img]
    for tag in tags:
        if tag in count_tags:
            count_tags[tag] += 1
        else:
            count_tags[tag] = 1
n_doc = len(train_tags)
idf = {t: math.log(n_doc / (1 + count_tags[t])) for t in count_tags}
```

```python
idf
```

```python
tf_idf = {}
for img in train_tags:
    tf_idf[img] = sum([idf[t] for t in train_tags[img]])
```

```python
sum_tf = sum(tf_idf.values())
proba = {img: tf_idf[img] / sum_tf for img in tf_idf}
```
